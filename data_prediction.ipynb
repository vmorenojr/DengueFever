```python
from io import StringIO
from csv import writer 
import pandas as pd
dhab = pd.read_csv('./Dados/dengue_por_habitante.csv.gz')
dhab = dhab[dhab['dt_sintoma']>= '2007'].reset_index()
dhab.index = dhab.UF
data_hoje_l = list(set(dhab[dhab['dt_sintoma']>='2007'].loc['RS'].dt_sintoma))
data_hoje_l.sort()
dist = pd.read_csv('distancia_UF.csv', index_col=0)
data_hoje1 = list(set(dhab[dhab['dt_sintoma']>='2010'].loc['RS'].dt_sintoma))
data_hoje1.sort()
```

```python
#Essa lista compreende as datas onde começaremos a analisar as cidades, pois a prinípio voltaremos 3 anos para cada cidade,
#ou seja, devemos começar a partir de 2010

data_hoje1
```

```python
output = StringIO() #Cria uma instância de string na memória
csv_writer = writer(output) # Permite escrever em csv nessa string. Tudo acontece na memória RAM, como o trabalho é apenas
                            # inserir strings (1 byte) na memória RAM, isso é muito mais rápido do que trabalhar diretamente
                            # com o pandas, pois o tipo de dado DataFrame é custoso e o pandas muitas vezes precisa realocar
                            # o DataFrame todo para uma posição na memória RAM onde contenha mais espaço vazio, para adicionar
                            # novos dados. Por isso a inserção de dados no pandas fica cada vez mais lenta. Pelo que li,
                            # esse método é 1000x mais rápido que com o pandas e não fica mais lento com o tempo.

# Quando o pandas lê o csv, a primeira linha do arquivo é o que ele naturalmente coloca como o nome das colunas.
# Por isso, inicialmente já escrevo as colunas no output
csv_writer.writerow(['cidade_alvo','data_hoje',
                     'ocorrencia_hoje','janela_ano',
                     'date_janela','ocorrencia_janela',
                     'cidade_pred','distancia',
                     'lag_semana','date_lag',
                     'ocorrencia_pred'])


#O valor que ele imprime é a posição do ponteiro na memória. Como ele imprimiu 138, significa que o arquivo tem 137 chars.
```

```python
for cidade_alvo in set(dhab.index): # começo fixando a cidade alvo
    for data_hoje in data_hoje1: # fixo a data que analisaremos com respeito à cidade alvo
        for janela in range(1,4): # aqui fixo a janela, de 1 a 3 anos
            for cidade_pred in set(dhab.index): # percorre as cidades que serão as preditoras (não exclui a prpópria cidade, mas excluir pode poupar algumas horas)
                for lag_pred in range(0,157): # agora percorro o lag da cidade preditora, de 0 semanas até 156 semanas
                    janelaData = data_hoje_l[data_hoje_l.index(data_hoje)-52*janela] # aqui guardo a data referente à janela que existe na planilha, 
                                                                                     # voltando 52*janela índices na lista, ou seja 52*janela semanas
                    lagData = data_hoje_l[data_hoje_l.index(data_hoje)-lag_pred] # mesma coisa, só que com a data do lag
                    # a row é escrita como uma lista em python, na ordem das colunas que existe no csv:
                    row = [cidade_alvo,
                           data_hoje,
                           dhab[dhab['dt_sintoma']==data_hoje].loc[cidade_alvo]['ocorrencias'],
                           janela,
                           janelaData,
                           dhab[dhab['dt_sintoma']==janelaData].loc[cidade_alvo]['ocorrencias'],
                           cidade_pred,
                           dist.loc[cidade_alvo,cidade_pred],
                           lag_pred,
                           lagData,
                           dhab[dhab['dt_sintoma']==lagData].loc[cidade_pred]['ocorrencias']]
                    csv_writer.writerow(row)
```

```python
# Quando o trabalho termina, o ponteiro de leitura está no final da string, portanto precisamos retornar
# ao primeiro char para então ler com o pandas e depois salvar no HD.

output.seek(0)
df = pd.read_csv(output)
df.to_csv('dados_final.csv')
```
